In Pearson correlation we calculate relation between each and every feature and we check 2 factors i.e, Direction of Relationship and the strength in that relationship(Covariance helps to find this)

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
df = pd.DataFrame(data.data)
```

```python
df
```

```python
corr_ = df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(18,18))
sns.heatmap(corr_,annot=True)
plt.show()
```

```python
def correlated_features(dataset, threshold):
    ans = set()
    corr_ = dataset.corr()
    for i in range(len(corr_)):
        for j in range(i):
            if abs(corr_.iloc[i,j]) > threshold:
                col_name = dataset.columns[i]
                ans.add(col_name)
    return ans
```

```python
cor_fet = correlated_features(dataset=df, threshold=0.7)
```

```python
df.drop(cor_fet,axis=1)
```
